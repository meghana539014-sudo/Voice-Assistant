```python
from gtts import gTTS
from IPython.display import Audio, display, Javascript
import datetime
from zoneinfo import ZoneInfo
import wikipedia
import smtplib
from email.message import EmailMessage
from urllib.parse import quote_plus
import time
import os

email_dict = {
    "friend": "friend@example.com",
    "family": "family@example.com"
}

audio_count = 0

def speak(text):
    global audio_count

    print("Assistant:", text)

    audio_count += 1
    filename = f"voice_{audio_count}.mp3"

    tts = gTTS(text=text, lang="en")
    tts.save(filename)

    display(Audio(filename, autoplay=True))

    # Wait until current audio finishes before next audio starts
    wait_time = max(2, len(text) * 0.08)
    time.sleep(wait_time)

def open_link(url):
    display(Javascript(f'window.open("{url}", "_blank");'))

def wishMe():
    hour = datetime.datetime.now(ZoneInfo("Asia/Kolkata")).hour

    if 0 <= hour < 12:
        speak("Good Morning!")
    elif 12 <= hour < 18:
        speak("Good Afternoon!")
    else:
        speak("Good Evening!")

    speak("I am Virusta. Please tell me how may I help you")

def sendEmail(to, content):
    sender_email = "your_email@gmail.com"
    app_password = "your_gmail_app_password"

    msg = EmailMessage()
    msg["From"] = sender_email
    msg["To"] = to
    msg["Subject"] = "Message from Virusta"
    msg.set_content(content)

    server = smtplib.SMTP_SSL("smtp.gmail.com", 465)
    server.login(sender_email, app_password)
    server.send_message(msg)
    server.quit()

wishMe()

while True:
    query = input("\nEnter your command: ").lower().strip()

    if query == "exit":
        speak("Goodbye!")
        break

    elif "wikipedia" in query:
        try:
            speak("Searching Wikipedia...")
            search_query = query.replace("wikipedia", "").strip()

            if search_query == "":
                speak("Please enter what you want to search on Wikipedia.")
            else:
                results = wikipedia.summary(search_query, sentences=2)
                print(results)
                speak(results)

        except Exception as e:
            print(e)
            speak("Sorry, I could not find results on Wikipedia.")

    elif "open youtube" in query:
        speak("Opening YouTube")
        open_link("https://www.youtube.com")

    elif "open google" in query:
        speak("Opening Google")
        open_link("https://www.google.com")

    elif "play music" in query:
        speak("Opening music")
        open_link("https://music.youtube.com")

    elif "the time" in query:
        strTime = datetime.datetime.now(ZoneInfo("Asia/Kolkata")).strftime("%I:%M:%S %p")
        speak(f"The current time is {strTime}")

    elif "search google for" in query:
        search_query = query.replace("search google for", "").strip()

        if search_query == "":
            speak("Please enter what you want to search on Google.")
        else:
            encoded_query = quote_plus(search_query)
            speak(f"Searching Google for {search_query}")
            open_link(f"https://www.google.com/search?q={encoded_query}")

    elif "search youtube for" in query:
        search_query = query.replace("search youtube for", "").strip()

        if search_query == "":
            speak("Please enter what you want to search on YouTube.")
        else:
            encoded_query = quote_plus(search_query)
            speak(f"Searching YouTube for {search_query}")
            open_link(f"https://www.youtube.com/results?search_query={encoded_query}")

    elif "search in hindi for" in query:
        search_query = query.replace("search in hindi for", "").strip()

        if search_query == "":
            speak("Please enter what you want to search in Hindi.")
        else:
            encoded_query = quote_plus(search_query)
            speak(f"Searching in Hindi for {search_query}")
            open_link(f"https://www.google.com/search?hl=hi&q={encoded_query}")

    elif "search in gujarati for" in query:
        search_query = query.replace("search in gujarati for", "").strip()

        if search_query == "":
            speak("Please enter what you want to search in Gujarati.")
        else:
            encoded_query = quote_plus(search_query)
            speak(f"Searching in Gujarati for {search_query}")
            open_link(f"https://www.google.com/search?hl=gu&q={encoded_query}")

    elif "send email to" in query:
        try:
            content = input("What should I say? ")
            recipient = input("Who should I send it to? ").lower().strip()

            if recipient in email_dict:
                to = email_dict[recipient]
                sendEmail(to, content)
                speak("Email has been sent!")
            else:
                speak("Recipient not found in email dictionary.")

        except Exception as e:
            print(e)
            speak("Sorry, I am not able to send this email.")

    else:
        speak("Sorry, I did not understand that command.")
```
